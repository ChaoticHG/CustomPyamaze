```python
import datetime
import sys
from pyamazeM import COLOR, maze, agent
import iterativeaStar
import aStar
```

```python
m1 = maze(sets=0)
m1runtime =[]
m1.CreateMaze(loadMaze="maze1.csv")
start = (15,15)
goal = (2,2)
threshold = iterativeaStar.h(start,goal)
start_time1 = datetime.datetime.now()
path1=aStar.aStar(m1,start,goal)
execution_time1 = datetime.datetime.now() - start_time1
start_time2 = datetime.datetime.now()
path2=iterativeaStar.iterative_deepening_a_star(m1,start,goal,threshold)
execution_time2 = datetime.datetime.now() - start_time2
execution_time1ms = execution_time1.total_seconds()
execution_time2ms = execution_time2.total_seconds()
a=agent(m1,15,15,footprints=True,color=COLOR.red)
b=agent(m1,15,15,footprints=True)
m1.tracePath({a:path1,b:path2},goal,kill=True,delay=300)
m1.setGoal1(2,2)
m1runtime.append(execution_time1ms)
m1runtime.append(execution_time2ms)
m1.run()


```

```python
import matplotlib.pyplot as plt
import numpy as np
  
allrun =[[None for i in range(6)] for j in range(5)] 
allrun[0][0]=m1runtime[0]
allrun[0][1]=m1runtime[1]
for x in range(4):
    allrun[0][x+2]=m1.runtime[x]
    
allrun[1][0]=m2runtime[0]
allrun[1][1]=m2runtime[1]
for x in range(4):
    allrun[1][x+2]=m2.runtime[x]
    
allrun[2][0]=m3runtime[0]
allrun[2][1]=m3runtime[1]
for x in range(4):
    allrun[2][x+2]=m3.runtime[x]
    
allrun[3][0]=m4runtime[0]
allrun[3][1]=m4runtime[1]
for x in range(4):
    allrun[3][x+2]=m4.runtime[x]
    
allrun[4][0]=m5runtime[0]
allrun[4][1]=m5runtime[1]
for x in range(4):
    allrun[4][x+2]=m5.runtime[x]
allrun1=[row[0] for row in allrun]
allrun2=[row[1] for row in allrun]
allrun3=[row[2] for row in allrun]
allrun4=[row[3] for row in allrun]
allrun5=[row[4] for row in allrun]
allrun6=[row[5] for row in allrun]
```

```python
labels = ['M1', 'M2', 'M3', 'GM4', 'M5']
x = np.arange(len(labels))
width = 0.15
fig, ax = plt.subplots(figsize=(12, 8))
rects1 = ax.bar(x , allrun1, width, label='Agent 1 Path 1')
rects2 = ax.bar(x + width, allrun2, width, label='Agent 2 Path 1')
rects3 = ax.bar(x + width*2, allrun3, width, label='Agent 1 Path 2')
rects4 = ax.bar(x + width*3, allrun4, width, label='Agent 2 Path 2')
rects5 = ax.bar(x + width*4, allrun5, width, label='Agent 1 Path 3')
rects6 = ax.bar(x + width*5, allrun6, width, label='Agent 2 Path 3')

ax.set_ylabel('Run Time(ms)')
ax.set_title('Total Run Time')
ax.set_xticks(x, labels)
ax.legend()
ax.bar_label(rects1, padding=5)
ax.bar_label(rects2, padding=5)
ax.bar_label(rects3, padding=5)
ax.bar_label(rects4, padding=5)
ax.bar_label(rects5, padding=5)
ax.bar_label(rects6, padding=5)

fig.tight_layout()

plt.show()
```
